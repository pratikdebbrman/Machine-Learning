# To predict diabetes using PIMA diabetes data

### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv("pima-data.csv")
```

```python
data.head(5)
```

```python
# check if any null value is present
data.isnull().values.any()
```

```python
data.dtypes
```

```python
data['num_preg'] = data.num_preg.astype(float)
data['glucose_conc'] = data.glucose_conc.astype(float)
data['diastolic_bp'] = data.diastolic_bp.astype(float)
data['thickness'] = data.thickness.astype(float)
data['insulin'] = data.insulin.astype(float)
data['age'] = data.age.astype(float)
data['diabetes'] = data.diabetes.astype(float)
```

## Changing the diabetes column data from boolean to number

```python
def replace_diabetes(my_string):
    if my_string == True:
        return 1
    elif my_string == False:
        return 0
```

```python
data['diabetes'] = data['diabetes'].apply(replace_diabetes)
```

```python
data.isnull().sum()
```

```python
x=data.drop(['diabetes'],axis=1)
y=data['diabetes']
```

```python
from sklearn.impute import SimpleImputer 
imputer=SimpleImputer(missing_values=0.0,strategy='mean')
imputer=imputer.fit(data.iloc[:,1:9])
data.iloc[:,1:9]=imputer.transform(data.iloc[:,1:9])
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(x,y,test_size = 0.3,random_state=56)
```
