```
dataset['Applicant_Gender'].replace({None:'M'}, inplace=True)
dataset['Applicant_Marital_Status'].replace({None:'M'}, inplace=True)
dataset['Applicant_Occupation'].replace({None:'Salaried'}, inplace=True)
dataset['Applicant_Qualification'].replace({None:'Class XII'}, inplace=True)
```

```
dataset=pd.get_dummies(dataset)
```

```
x=dataset.drop(['Business_Sourced'],axis=1)
y=dataset['Business_Sourced']
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```
from sklearn.metrics import auc,f1_score,roc_auc_score
predict_train=lr.predict_proba(x_train)
predict_test=lr.predict_proba(x_test)
```

```
roc_auc_score(y_train,predict_train[:,1])
```

```
roc_auc_score(y_test,predict_test[:,1])
```

```
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=11, max_depth= 5, random_state=1)
rfc.fit(x_train,y_train)
```

```
predict_train_forest=rfc.predict_proba(x_train)
predict_test_forest=rfc.predict_proba(x_test)
```

```
roc_auc_score(y_train,predict_train_forest[:,1])
```

```
roc_auc_score(y_test,predict_test_forest[:,1])
```
