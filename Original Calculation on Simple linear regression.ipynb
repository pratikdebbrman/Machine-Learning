Original Calculation on Simple linear regression

```python

import numpy as np

# x,y=int(input()),int(input())

# B1 = sum((x(i) - mean(x)) * (y(i) - mean(y))) / sum( (x(i) - mean(x))^2 )
# B0 = mean(y) - B1 * mean(x)
 
x,y=[1,2,4,3,5],[1,3,3,2,5]

# total_corrdinates=int(input("Enter total no of corrdinates ")) 
# for i in range(0,total_corrdinates):
#   x.append(int(input("Enter X"+str(i)+" ")))
#   y.append(int(input("Enter Y"+str(i)+" ")))

sum1,sum2=0,0
for i in range(0,len(x)):
  sum1+=((x[i] - np.mean(x)) * (y[i] - np.mean(y))) 
  sum2+=((x[i] - np.mean(x))**2)
B1 = sum1 / sum2
B0 = np.mean(y) - B1 * np.mean(x)

for i in range(len(x)):
  final_y= B0 + B1 * x[i]
  print("The following predictions for ("+str(x[i])+","+str(y[i])+") is "+str(final_y))
```

Original Calculation on Simple linear regression and ploting the graph

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_regression_line(x, y, b):

	# plotting the actual points as scatter plot
  plt.scatter(x, y, color = "m",marker = "o", s = 30)


	# predicted response vector
  y_pred = b[0] + b[1]*np.array(x)

  print("The Original Value of X is "+str(np.array(x)))
  print("The Original Value of Y is "+str(np.array(y)))
  print("The Predicted Value is "+str(y_pred))
	# plotting the regression line
  plt.plot(x, y_pred, color = "g")

	# putting labels
  plt.xlabel('x')
  plt.ylabel('y')

	# function to show plot
  plt.show()

# Below is the original Simple Linear Regression calculation    
x,y=[1,2,4,3,5],[1,3,3,2,5]

# total_corrdinates=int(input("Enter total no of corrdinates "))
# for i in range(0,total_corrdinates):
#   x.append(int(input("Enter X"+str(i)+" ")))
#   y.append(int(input("Enter Y"+str(i)+" ")))


sum1,sum2=0,0
for i in range(0,len(x)):
    # Calculate covariance between x and y
    sum1+=((x[i] - np.mean(x)) * (y[i] - np.mean(y))) 
    # Calculate the variance of a list of numbers
    sum2+=((x[i] - np.mean(x))**2)

# 	y = B0 + B1 * x
B1 = sum1 / sum2   
B0 = np.mean(y) - B1 * np.mean(x)

# for i in range(len(x)):
#     final_y= B0 + B1 * x[i]
#     print(final_y)

# plotting regression line
b = B0, B1
plot_regression_line(x, y, b)
```
